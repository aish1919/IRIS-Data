```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3, 
                                                random_state=1)
```

```python
#from tensorflow.keras.callbacks import EarlyStopping
#early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()

import tensorflow as tf
model.add(Dense(20,activation='relu')) #Input layer (Number of features - 1(label))
model.add(Dense(20,activation='relu'))
model.add(Dense(3,activation='softmax')) #Output layer (Since it's a binary classification problem)

#Using accuracy as loss function
model.compile(optimizer='adam',loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

model.fit(xtrain,ytrain,epochs=200,validation_data=(xtest, ytest),verbose=1,batch_size=128)
```

```python
model.history.history
```
